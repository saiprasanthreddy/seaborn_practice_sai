```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
df.tail()
```

```python
df['smoker'].value_counts()
```

```python
df.dtypes
```

```python
c= df.corr(numeric_only=True)
c
```

```python
sns.heatmap(c, cmap='coolwarm')
```

```python
sns.jointplot(x='tip', y='total_bill',data= df,kind='hex' )
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.displot(df['total_bill'],kde=True, bins = 50)
```

```python
sns.countplot(x= 'sex', data= df)
```

```python
sns.countplot(x= 'smoker', data= df, hue= 'sex')
```

```python
sns.boxplot(x='sex', y='total_bill', data=df , hue='smoker', palette='rainbow')
```

```python
sns.violinplot(x='sex', y='total_bill', data=df , hue='smoker', palette='rainbow')
```
